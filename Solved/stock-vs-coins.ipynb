```python
bitcoin_vs_ethereum_df=pd.read_sql_query('SELECT * FROM bitcoin INNER JOIN ethereum ON bitcoin.Date=ethereum.Date', con=engine)
```

```python
###----note: un-command the following line for debugging purpose ------------
#
#   bitcoin_vs_ethereum_df
#
####------end of note --------
bitcoin_vs_ethereum_df.head()
```

```python
# Set x axis to Date
x_axis_data = bitcoin_vs_ethereum_df.iloc[:,0].tolist()
x_axis_data.reverse()

# Set y axis to Price
points = bitcoin_vs_ethereum_df.iloc[:,1].tolist()
points.reverse()
points_C = bitcoin_vs_ethereum_df.iloc[:,5].tolist()
points_C.reverse()


###---------------- For debugging purposes------------
#print(x_axis_data)
#print(len(points))

# Incorporate the other graph properties
plt.title("year2014-2018: Bitcoin Price vs Ethereum Price")
plt.ylabel("Closing Price ($)")
plt.xlabel("Date")


# # Create a legend
#plt.legend()

# Incorporate a text label
plt.text(42, 35, "Note:BitCoin in Blue; Ethereum in Orange")

# Plot the line
plt.figure(1)
plt.plot(x_axis_data, points)
plt.plot(x_axis_data, points_C)
plt.show()

### Save Figure 1
plt.savefig("../Images/bitcoin_vs_ethereum.png")
```

```python
ibm_vs_msft_df=pd.read_sql_query('SELECT * FROM ibm INNER JOIN msft ON ibm.date=msft.date', con=engine)
```

```python
###----note: un-command the following line for debugging purpose ------------
#
#   ibm_vs_msft_df
#
####------end of note --------
ibm_vs_msft_df.head()
```

```python
# Set x axis to Date
x_axis_data_2 = ibm_vs_msft_df.iloc[:,0].tolist()

# Set y axis to Price
points_2 = ibm_vs_msft_df.iloc[:,1].tolist()
points_D = ibm_vs_msft_df.iloc[:,4].tolist()

###---------------- For debugging purposes------------
#print(x_axis_data)
#print(len(points))

# Plot the line
plt.figure(2)
plt.plot(x_axis_data_2, points_2)
plt.plot(x_axis_data_2, points_D)
plt.show()

# Incorporate the other graph properties
plt.title("year2014-2018: IBM Price vs MSFT Price")
plt.ylabel("Closing Price ($)")
plt.xlabel("Date")

# Incorporate a text label
plt.text(42, 35, "Note: IBM in Blue; MSFT in Orange")

### Save Figure 1
plt.savefig("../Images/ibm_vs_msft.png")
```
